```python
#function to get sorted values for the correlation between the features 
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(df):
    au_corr = df.corr().abs().stack()
    labels_to_drop = get_redundant_pairs(df.corr())
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr
```

```python
#splitting the data as test and train
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size= 0.33, random_state=123)
```

```python
#function to get rmse values
def get_rmse(data, target, cmat):
    dlength = len(data)
    loss = np.dot(data,cmat)- target
    return np.sqrt(np.sum(np.power(loss,2))/dlength)
```

```python
#function to get the cost
def get_cost(data, target, cmat):
    dlength = len(data)
    loss = np.dot(data,cmat)- target
    return np.sum(np.power(loss,2))/dlength
```

```python
#function to perform the gradient descent
def get_theta_matrix(alpha, iterations, data, target, cmat):
    dlength = len(data)
    for i in range (0, iterations):
        loss = np.dot(data,cmat)- target
        gradient = pd.DataFrame(np.dot(data.transpose(),loss)).divide(dlength)
        cmat = cmat - (gradient.multiply(alpha))
    return cmat
```

```python
#function to run the linear regression for varies learning rates and iteration.
def run_linear_regress(data, target, cmat):
    alpha_values = [0.75,0.5,0.25,0.1]
    iteration = [500,1000,2500,5000, 7500, 10000]
    theta_matrixes = {}
    rmse_values = {}
    cost_values = {}
    for i in alpha_values:
        for j in iteration:
            theta_matrixes[str(i),str(j)] = get_theta_matrix(i,j,data,target,cmat)
            rmse_values[str(i),str(j)] = get_rmse(data, target, theta_matrixes[str(i),str(j)])
            cost_values[str(i),str(j)] = get_cost(data, target, theta_matrixes[str(i),str(j)])
    return theta_matrixes,rmse_values,cost_values
```

```python
#creating an empty coefficient matrix
carray = np.zeros(len(X_train.columns))
cmat = pd.DataFrame(carray)
cmat.shape
```

```python
#feature scaling
scaler = MinMaxScaler(feature_range=(0, 1))
X_Scalar_train = scaler.fit_transform(X_train)
y_Scalar_train = scaler.fit_transform(y_train)
```

```python
#run linear regression with the train set
thetas,rmse_val,cost_val = run_linear_regress(X_Scalar_train,y_Scalar_train, cmat)
```
